```python
# Importing packages.
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science', 'notebook', 'grid'])

# Defining the simulation function.
def simulation(n):
    '''
    Input:
    n: int
        The number of (x,y) pairs of the simulation.
    Output:
    p: float
        The resulting experimental probability.
    '''
    X = np.random.rand(n) # Creating n number of delays (Romeo)
    Y = np.random.rand(n) # Creating n number of delays (Juliet)
    z = np.zeros(n) # Matrix for counting the "meetings"
    num_of_meeting_points = len(z[np.abs(X-Y)<=0.25]) # Counting the "meetings"
    p = num_of_meeting_points/n # Calculating possibility.
    return p

n_points = [] # Storing number of n pairs.
prob = [] # Storing probability approximations.
for i in np.logspace(1,8,12): # Simulating for multiple number of pairs.
    i = int(i)
    n_points.append(i) # Storing n values.
    prob.append(simulation(i)) # Storing the calculated probabilities.
    
# Plotting the convergence of the method.
plt.plot(n_points,prob, 'ko-')
plt.xscale('log')
plt.axhline(7/16, color='red', linestyle='--', linewidth=1, label='Theoretical value')
plt.xlabel('Number of (x,y) pairs')
plt.ylabel('"Success" percentage')
plt.title('Convergence of simulation')
plt.legend(frameon=True, edgecolor='black', fontsize=11)
#plt.savefig('convergence.png')
plt.show()
```

```python
import random

def simulate_meeting_probability(num_trials=1000000, wait_time=0.25):
    """
    Simulate the probability that Romeo and Juliet meet given:
    - Both arrive randomly between 0 and 1 hour.
    - Each waits 15 minutes (0.25 hours) for the other.
    
    Args:
        num_trials (int): Number of simulations to run.
        wait_time (float): Time (in hours) each person waits.

    Returns:
        float: Estimated probability of meeting.
    """
    meet_count = 0

    for _ in range(num_trials):
        romeo_arrival = random.uniform(0, 1)
        juliet_arrival = random.uniform(0, 1)

        # They meet if the time difference is within the waiting time
        if abs(romeo_arrival - juliet_arrival) <= wait_time:
            meet_count += 1

    return meet_count / num_trials

# Run the simulation
estimated_probability = simulate_meeting_probability()
print(f"Estimated probability of meeting: {estimated_probability:.5f}")
```
